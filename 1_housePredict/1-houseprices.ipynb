```python
import pandas as pd

train_ds = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
train_ds
```

```python
test_ds = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv')
test_ds
```

```python
train_ds['SalePrice'].mean()
```

```python
test_ds['SalePrice'] = train_ds['SalePrice'].mean()
test_ds
```

```python
test_ds[['Id', 'SalePrice']].to_csv('house_predict_submission_01.csv', index=False)
```

```python
from copy import deepcopy

train_ds_sub = deepcopy(train_ds[['MSSubClass', 'LotArea', 'SalePrice']])
train_ds_sub = train_ds_sub.fillna(train_ds_sub.mean())
train_ds_sub
```

```python
x_train = train_ds_sub.to_numpy()[:, :2]
y_train = train_ds_sub.to_numpy()[:, 2:]
print(x_train)
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(x_train, y_train)
lr
```

```python
# this will show the beta value(s) like, b1, b2 etc
lr.coef_
```

```python
# this is the b0 of linear regression equation
lr.intercept_
```

```python
train_ds_sub['SalePricePredict'] = lr.predict(x_train)
train_ds_sub
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(train_ds_sub['SalePrice'], train_ds_sub['SalePricePredict'])
```

```python
test_ds_sub = test_ds[['MSSubClass', 'LotArea']]
test_ds_sub = test_ds_sub.fillna(test_ds_sub.mean())
x_test = test_ds_sub.to_numpy()
y_predict = lr.predict(x_test)
test_ds_sub['SalePrice'] = y_predict
test_ds_sub['Id'] = test_ds[['Id']]
test_ds_sub[['Id', 'SalePrice']].to_csv('linearReg_Predict.csv', index=False)
```
